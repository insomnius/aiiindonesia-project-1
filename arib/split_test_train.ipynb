# Split Train Test

IPYNB ini digunakan untuk memisahkan train dan test.

```python
import pandas as pd
import os
import shutil
```

```python
# Directory paths
directories = ['test/', 'train/']

# Function to remove all files in a directory
def remove_files(directory):
    for file_name in os.listdir(directory):
        file_path = os.path.join(directory, file_name)
        if os.path.isfile(file_path):
            os.remove(file_path)
            print(f"Removed file: {file_path}")

# Remove files from each directory
for directory in directories:
    if os.path.exists(directory):
        remove_files(directory)
    else:
        print(f"Directory does not exist: {directory}")
```

```python
df = pd.read_csv('list_attribute.txt', delim_whitespace=True, index_col=0)
image_files = os.listdir('images/')
filtered_df = df[df.index.isin(image_files)]
filtered_df
```

```python
print("Male count:", filtered_df['Male'].value_counts()[1])
print("Female count:", filtered_df['Male'].value_counts()[-1])
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(filtered_df, test_size=0.3)
```
